```python
import pandas as pd
import matplotlib.pyplot as plt
import itertools
%matplotlib inline
```

```python
women_degrees = pd.read_csv('C:\\Users\\yeachan153\\Desktop\\BehaviouralDataScience\\Python\\Datasets\\percent-bachelors-degrees-women-usa.csv')
```

```python
# The data refers to the proportion of genders in each field of study (colour blind friendly)

subjects = ['Biology', 'Computer Science', 'Engineering', 'Math and Statistics']
spines = ['right', 'left', 'bottom', 'top']
fig = plt.figure(figsize = (10,10))
for i in range(4):   
    ax = fig.add_subplot(2,2,i+1)
    ax.plot(women_degrees['Year'], women_degrees[subjects[i]],linewidth = 3, label = 'Women', color = (0/255, 107/255,164/255))
    ax.plot(women_degrees['Year'], 100-women_degrees[subjects[i]], linewidth = 3, label = 'Men', color = (255/255,128/255,14/255))
    ax.tick_params(bottom = 'off', top = 'off', left = 'off', right = 'off')
    ax.set_title(subjects[i])
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    if i == 3:
        plt.legend(loc = 'upper right')
    for each in spines:
        ax.spines[each].set_visible(False)
```

We can see that more men studied Biology than women until the 1990's, however recently more women study it. Proportionally, more women seemed to be getting into computer science from the 1970's until the early/mid 1980's, however this trend abruptly stopped. It is hard to tell exactly what is happening. It could be the case that there was a sudden influx of male computer scientists from the 1980's, or the number of women who were applying did not grow any further whole the number of men applying did, or the number of women applying grew at the same rate, however the number of men applying grew exponentially. In engineering, there is a very big gender gap that seems to be narrowing. The gender gap in math and statistics is relatively small, with the number of men studying proportionally higher.

```python
# Graphs can be ordered (highest recent gender gap - lowest recent gender gap) for easier viewing

subjects = ['Engineering', 'Computer Science', 'Psychology', 'Biology', 'Physical Sciences', 'Math and Statistics']
spines = ['right', 'left', 'bottom', 'top']
fig = plt.figure(figsize = (18,3))
for i in range(6):   
    ax = fig.add_subplot(1,6,i+1)
    ax.plot(women_degrees['Year'], women_degrees[subjects[i]],linewidth = 3, label = 'Women', color = (0/255, 107/255,164/255))
    ax.plot(women_degrees['Year'], 100-women_degrees[subjects[i]], linewidth = 3, label = 'Men', color = (255/255,128/255,14/255))
    if i == 0:
        ax.text(x = 2005, y = 87, s='Men')
        ax.text(x = 2002, y = 8, s='Women')
    if i == 5:
        ax.text(x = 2005, y = 62,s= 'Men')
        ax.text(x = 2001, y = 35, s='Women')
    ax.tick_params(bottom = 'off', top = 'off', left = 'off', right = 'off')
    ax.set_title(subjects[i])
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    for each in spines:
        ax.spines[each].set_visible(False)
```

Ordering from current highest gendergap to current lowest gendergap makes it easier to spot patterns for the reader. For example, we can see that in the present year, engineering has the biggest gender gap in enrollment, whereas math and statistics has the smallest.

```python
# Showing a 6 x 3 figure, each column represents stem_cats, lib_arts_cats & other_cats

stem_cats = ['Psychology', 'Biology', 'Math and Statistics', 'Physical Sciences', 'Computer Science', 'Engineering']
lib_arts_cats = ['Foreign Languages', 'English', 'Communications and Journalism', 'Art and Performance', 'Social Sciences and History']
other_cats = ['Health Professions', 'Public Administration', 'Education', 'Agriculture','Business', 'Architecture']
subjects = []
for each_tuple in itertools.zip_longest(stem_cats, lib_arts_cats, other_cats):
    for subject in range(3):
        subjects.append(each_tuple[subject])
        
women_colour = (0/255, 107/255,164/255)
men_colour = (255/255,128/255,14/255)
colour_blind3 = (207/255, 207/255, 207/255)

fig = plt.figure(figsize = (18,22))
for subj_idx in range(18):
    if subj_idx != 16:
        ax = fig.add_subplot(6,3, subj_idx+1)
        ax.plot(women_degrees['Year'], women_degrees[subjects[subj_idx]], label = 'Women', color = women_colour, linewidth = 3)
        ax.plot(women_degrees['Year'], 100 - women_degrees[subjects[subj_idx]], label = 'Men', color = men_colour, linewidth = 3)
        ax.set_title(subjects[subj_idx])
        ax.set_ylim(0,100)
        ax.set_xlim(1970,2011)
        ax.tick_params(bottom = 'off', left = 'off', labelbottom = 'off')
        ax.set_yticks([0,100])
        ax.axhline(50, color = colour_blind3, alpha = .7)
    if (subj_idx == 0):
        ax.text(x = 2005, y = 82, s = 'Women')
        ax.text(x = 2007, y = 15, s = 'Men')
    elif (subj_idx ==1):
        ax.text(x = 2005, y = 75, s = 'Women')
        ax.text(x = 2007, y = 22, s = 'Men')
    elif (subj_idx == 2):
        ax.text(x = 2005, y = 90, s = 'Women')
        ax.text(x = 2007, y = 7, s = 'Men')            
    elif (subj_idx == 15):
        ax.text(x = 2005, y = 87, s = 'Women')
        ax.text(x = 2007, y = 9, s = 'Men') 
        ax.tick_params(labelbottom = 'On')
    elif (subj_idx == 17):
        ax.text(x = 2005, y = 65, s = 'Women')
        ax.text(x = 2007, y = 35, s = 'Men') 
        ax.tick_params(labelbottom = 'On')
```
